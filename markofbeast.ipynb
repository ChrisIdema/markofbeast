```python
import libscrc
import re

# https://aptparenting.com/common-first-names-is-your-name-in-list
#https://pypi.org/project/libscrc/
names = ['Chris','Christopher','Daniel','Paul','Mark','Donald',
'George','Kenneth','Steven','Edward','Brian',
'Ronald','Anthony','Kevin','Jason','Matthew',
'Gary','Timothy','Jose','Larry','Jeffrey',
'Frank','Scott','Eric','Stephen','Andrew',
'Raymond','Jack','Gregory','Joshua','Jerry',
'Dennis','Walter','Patrick','Peter','Harold',
'Douglas','Henry','Carl','Arthur','Ryan',
'Mary','Patricia','Barbara','Linda','Elizabeth',
'Maria','Jennifer','Susan','Margaret','Dorothy',
'Lisa','Nancy','Karen','Betty','Helen',
'Sandra','Donna','Ruth','Sharon','Michelle',
'Laura','Sarah','Kimberly','Deborah','Jessica',
'Shirley','Cynthia','Angela','Emily','Brenda',
'Amy','Anna','Rebecca','Virginia','Kathleen',
'Pamela','Martha','Debra','Amanda','Stephanie',
'Caroline','Christine','Marie','Janet','Catherine',
'Frances','Ann','Joyce','Diane','Alice']

for name in names:
    print(name)
    

    test = re.sub('[^aeiou]', "", name, flags=re.IGNORECASE)
    print(test)
    test2 = re.sub('[aeiou]', "", name, flags=re.IGNORECASE)
    print(test2)

    six_count = 0

    if len(name)==6:
        print('your name has 6 letters')
        six_count += 1

    digits = []

    for c in name.lower():
        digits.append(ord(c) - ord('a')+1)

#     print(digits)

    for i in range(len(name)):
        if digits[i]%6==0:            
            if digits[i]/6==3:
                print('letter #'+str(i+1)+' (numerical value ' + str(digits[i])+  ') is 6+6+6')
            else:
                print('letter #'+str(i+1)+' (numerical value ' + str(digits[i])+  ') is divisable by 6')

    # different  hashes, on 3 different cases
    # if it's equal to 666 or 6660 it should count

    crc_functions = [libscrc.xmodem,libscrc.modbus,libscrc.atm10,libscrc.gsm10,libscrc.crc10_cdma2000,libscrc.ibm,
                    libscrc.umts11,libscrc.flexray11,libscrc.umts12,libscrc.crc12_3gpp,libscrc.dect12,libscrc.crc12,
                    libscrc.gsm12,libscrc.crc12_cdma2000]
    
    
    for f in crc_functions:

        crc = f(name.encode())
        if crc in [6,66,666,6660]:
            print('The crc of your name is',crc)
            
        crc = f(name.lower().encode())
        if crc in [6,66,666,6660]:
            print('The crc of your name in lower case is',crc)

        crc = f(name.upper().encode())
        if crc in [6,66,666,6660]:
            print('The crc of your name in upper case is',crc)


        
        


```
